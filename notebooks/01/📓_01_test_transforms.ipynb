```python
from app.data_loader import load_with_sites

df = load_with_sites(kind="spatial", window="full")

df.head(), df.shape
```

```python
from app.transforms import aggregate_site_week

agg = aggregate_site_week(
    df,
    value_col="suitability_score",
    agg="mean"
)

agg.head()
```

```python
from app.transforms import build_site_week_matrix

matrix = build_site_week_matrix(
    agg,
    value_col="suitability_score"
)

matrix.shape
```

```python
from app.transforms import normalize_per_week

norm = normalize_per_week(matrix, method="minmax")

norm.min().min(), norm.max().max()
```

```python
from app.transforms import rank_per_week

ranks = rank_per_week(matrix, ascending=False)

ranks.iloc[:5, :5]
```
